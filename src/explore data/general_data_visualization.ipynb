# INSTALACIÓN E IMPORTACIÓN DE PAQUETERIAS NECESARIAS
Para poder realizar la extracción de metadata de forma automatica, se instalan las paqueterias necesarias para enseguida hacer la importación y que se encuentren disponibles para la ejecución del cuaderno.

### PREVIO A LA DESCARGA DE METADATA
Se necesita obtener tu propia Key de Kaggle, la obtienes de la siguiente manera:
1. Registrate para obtener una cuenta Kaggle en https://www.kaggle.com.
2. Da clic en tu foto de usuario, y da click en "Settings"
3. Selecciona "Create API token" o "Create New Token". 
Esto activará la descarga de kaggle.json, un archivo que contiene tus credenciales de API; este archivo lo deberas ubicar en el siguiente path: C:\Users\<usuario>\.kaggle

```python
%pip install kaggle
import os
from kaggle.api.kaggle_api_extended import KaggleApi
from pprint import pprint
```

## METADATA DEL DATASET 'BREAST CANCER PROTEOMES'
Para mostrar la metadata de este dataset, necesitas disponer tanto de tu usuario de Kaggle como tu propia Key, para poder ejecutar el siguiente bloque de código.

```python
os.environ['KAGGLE_USERNAME'] = "KAGGLE_USERNAME"
os.environ['KAGGLE_KEY'] = "KAGGLE_KEY"
owner = 'piotrgrabo'
datasetName = 'breastcancerproteomes'
api = KaggleApi()
api.authenticate()
metadata = api.metadata_get(owner, datasetName)
pprint(metadata)
```
