```python
import pandas as pd
import kaggle
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

kaggle_id = "piotrgrabo/breastcancerproteomes"
```

```python
# Download data from Kaggle and save it to 'data' folder
kaggle.api.authenticate()
kaggle.api.dataset_download_files(kaggle_id, path="./", unzip=True)

```

```python
df = pd.read_csv('77_cancer_proteomes_CPTAC_itraq.csv')
```

```python
df.head()
```

### Minor data cleaning

```python

# Drop non-numeric columns to keep only the expression data
df_numeric = df.drop(columns=['RefSeq_accession_number', 'gene_symbol', 'gene_name'])
```

## Visualization

```python

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric, cmap='viridis')
plt.title('Breast Cancer Proteome Heat Map')
plt.xlabel('Samples')
plt.ylabel('Genes')
plt.show()

```

```python
# Boxplot (for a specific gene as an example, you can change the index to visualize other genes)
gene_index = 0  # Change this index to visualize other genes
sns.boxplot(x=df_numeric.loc[gene_index])
plt.title(f'Box Plot of Expression Levels for Gene {df.gene_symbol[gene_index]}')
plt.xlabel('Expression Level')
plt.show()
```
