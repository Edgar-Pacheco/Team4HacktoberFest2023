FUNCIÓN PARA CREAR DB CON DUCKDB

## INICIALIZAR DB CON DUCKDB
Para poder crear las tablas con un formato simple, se debe de inicializar una DB mediante DuckDB, donde los datos solo seran llamados para mostrarse en tabla.

#### FUNCIÓN PARA SIMPLIFICAR CREACIÓN DE DB

```python
def dbcreate(url, archive):
    %load_ext sql
    %sql duckdb://
    from urllib.request import urlretrieve

    _ = urlretrieve(url, archive)
```

## CREACIÓN DE DB PARA DATASET 'BREAST CANCER PROTEOMES'

```python
dbcreate("https://raw.githubusercontent.com/Edgar-Pacheco/Team4HacktoberFest2023/explore-data/src/explore%20data/77_cancer_proteomes_CPTAC_itraq.csv",
         "77_cancer_proteomes_CPTAC_itraq.csv" )
```

#### CREACIÓN DE TABLA

```python
%%sql
CREATE TABLE kg_proteomes AS SELECT * FROM "77_cancer_proteomes_CPTAC_itraq.csv"
```

#### VISUALIZACIÓN DE LA TABLA

```python
%sqlcmd explore --table kg_proteomes
```
